```python
new_happiness=happy_df.drop(['Country','Region','Happiness Rank'],axis=1)
new_happiness.head()
new_happiness.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=new_happiness.drop(['Happiness Score'],axis=1).values
y=new_happiness['Happiness Score'].values
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
new_happiness=scale.fit_transform(new_happiness)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
y_pred=lm.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('predicted')
plt.title("Actual vs Predicted")
```

```python
#Predicted values
pred_y_df=pd.DataFrame({'Actual Value':y_test,'Predicted value':y_pred,'Difference':y_test-y_pred})
pred_y_df
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error as MSE
print('Root Mean Squared Error:',np.sqrt(MSE(y_test,y_pred)))
```
